# **Splitting the dataset into train and test set**

```python
# splitting into training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

**Applying the model**

```python
# apply simple linear regression model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

# y = w.x + b


# **Calculating coefficient,intercept**

```python
regressor.coef_ #w
```

```python
regressor.intercept_  #b
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred.shape
```

# **Scatter plot temp vs year**

```python
# plotting the scatter plot temp vs year
plt.scatter(X,y)
plt.plot(X, regressor.predict(X), color = 'red')
plt.title('Temperature vs Year')
plt.xlabel('Year')
plt.ylabel('Temperature')
plt.show()
```

```python
# regression plot
sns.regplot(x=X, y=y, ci=None, scatter_kws={'color':'b', 's':9})
```

# **calculating mean squared error, mean absolute error, r2-score**

```python
# calculating mean squared error, mean absolute error, r2-score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', mse)
print('Mean Absolute Error:', mae)
print('R2 Score:', r2)
```
