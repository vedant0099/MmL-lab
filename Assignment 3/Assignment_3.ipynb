```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/content/Heart.csv')
```

```python
dataset.describe()
```

```python
#type of each column
dataset.dtypes
```

# **Pie Chart**

```python
# plotting pie chart with attributes chestPain, RestBP, Chol, Fbs
#distribution of chestpain using pie chart
dataset.ChestPain.value_counts().plot.pie(autopct='%.2f')

plt.show()
```

# **Bar Plot**

Result:
Asymptomatic ChestPain type has highest and typical type of ChestPain is lowest.

```python
# distribution of chestPain type
# using bar plot
sns.countplot(x='ChestPain', data=dataset)
plt.title('Distribution of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

Result:
From above distribution we can conclude that the higher to lower level of ChestPain type as follows :-
asymptomatic > nonanginal > nontypical > typical

```python
# average cholesterol by AHD
sns.barplot(x='AHD', y='Chol', data=dataset, estimator=np.mean)
plt.title('Average Cholesterol Levels by AHD Status')
plt.xlabel('AHD (Presence/Absence)')
plt.ylabel('Average Cholesterol')
plt.show()
```

Result:-
In the presence of AHD, the cholesterol level is slightly higher

# **Line Chart**

```python
# line chart of age and chol
sns.lineplot(x='Age', y='Chol', data=dataset)
plt.title('Age vs Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```

Result:-
As age increases cholesterol level also increases


# **Box Plot**

```python
sns.boxplot(x='AHD', y='Chol', data=dataset)
plt.title('Boxplot of Cholesterol Levels by AHD Status')
plt.xlabel('AHD (Presence/Absence)')
plt.ylabel('Cholesterol')
plt.show()

```

Result:-
In the presence of AHD, cholesterol level is slightly higher

```python
sns.boxplot(x='ChestPain', y='Chol', data=dataset)
plt.title('Boxplot of Cholesterol Levels by ChestPain')
plt.xlabel('ChestPain')
plt.ylabel('Cholesterol')
plt.show()
```

Result:-
In asymptomatic ChestPain, cholesterol level is slightly higher

# **Scatter Plot**

```python
# scatter plot using age and MaxHR
sns.scatterplot(x='Age', y='MaxHR', data=dataset)
plt.title('Age vs Maximum Heart Rate')
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate')
plt.show()
```

Result:-
Heart rate is maximum between the age group 50 to 60

```python
# scatter plot using age and RestBP
sns.scatterplot(x='Age', y='RestBP', data=dataset)
plt.title('Age vs Resting Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.show()
```
